```python
import re
import ast
def convert(df):
    l=[]
    for i in df["product_category_tree"]:
        i=str(i)
        i=re.sub(r'\s\>\>\s',",",i)
        l.append([re.sub('\[|\]|\"|\'',"",i)])
    return l
#df["product_category_tree"].iloc[0]
```

```python
df["product_category_tree"]=convert(df)
```

```python
df["description"]=df["description"].apply(lambda x:x.split())
df["product_rating"]=df["product_rating"].apply(lambda x : x.split())
df["brand"]=df["brand"].apply(lambda x: x.split())
```

```python
df.head(1)
```

```python
df["product_category_tree"]=df["product_category_tree"].apply(lambda x: [i.replace(" ","") for i in x])
df["description"]=df["description"].apply(lambda x: [i.replace(" ","") for i in x])
df["product_rating"]=df["product_rating"].apply(lambda x: [i.replace(" ","") for i in x])
df["brand"]=df["brand"].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
df.head(1)
```

```python
df["tags"]=df["product_category_tree"]+df["description"]+df["product_rating"]+df["brand"]
```

```python
df["tags"]=df["tags"].apply(lambda x : " ".join(x))
df["tags"]=df["tags"].apply(lambda x : x.lower())
df["tags"]=df["tags"].apply(lambda x : x.replace(","," "))
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df["tags"]=df["tags"].apply(stem)
```

```python
df["tags"][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vector=cv.fit_transform(df["tags"]).toarray()
```

```python
vector.shape
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim=cosine_similarity(vector,vector)
```

```python
sim.shape
```

```python
sim[0]
```

```python
def recommend(product):
    product_index=df[df["product_name"]==product].index[0]
    distances=sim[product_index]
    product_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[0]
    id_= df.iloc[product_list[0]]["uniq_id"]
    return data[data["uniq_id"]==id_][["Product name in Flipkart","Retail Price in Flipkart","Discounted Price in Flipkart","Product name in Amazon","Retail Price in Amazon","Discounted Price in Amazon"]]
```
